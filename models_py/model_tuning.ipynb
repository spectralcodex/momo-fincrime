```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from collections import Counter

# Create a simple banking dataset
np.random.seed(42)
n_samples = 1000

# Generate features
age = np.random.randint(18, 70, n_samples)
income = np.random.randint(20000, 200000, n_samples)
loan_amount = np.random.randint(1000, 50000, n_samples)
credit_score = np.random.randint(300, 850, n_samples)

# Generate target (loan default)
# Let's assume 10% default rate
default = np.random.choice([0, 1], n_samples, p=[0.9, 0.1])

# Create DataFrame
df = pd.DataFrame({
    'age': age,
    'income': income,
    'loan_amount': loan_amount,
    'credit_score': credit_score,
    'default': default
})

# Separate features and target
X = df.drop('default', axis=1)
y = df['default']

# Print original class distribution
print("Original class distribution:", Counter(y))

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define sampling strategy
# We'll aim for a 50-50 split between default and non-default
n_samples_desired = 0.5
sampling_strategy = {0: n_samples_desired, 1: n_samples_desired}

# Create a pipeline with both over and under sampling
pipeline = Pipeline([
    ('over', RandomOverSampler(sampling_strategy=sampling_strategy, random_state=42)),
    ('under', RandomUnderSampler(sampling_strategy=sampling_strategy, random_state=42))
])
```

```python
over = RandomOverSampler(sampling_strategy=0.50, random_state=42)
X, y = over.fit_resample(X_scaled, y)
```

```python
over = RandomUnderSampler(sampling_strategy=0.5, random_state=42)
X, y = over.fit_resample(X_scaled, y)
```

```python
print("Original class distribution:", Counter(y))
```

```python

# Fit and transform the data
X_resampled, y_resampled = pipeline.fit_resample(X_scaled, y)

# Print new class distribution
print("Resampled class distribution:", Counter(y_resampled))

# Print shape of original and resampled data
print("Original shape:", X_scaled.shape)
print("Resampled shape:", X_resampled.shape)

# Convert resampled data back to DataFrame for easier viewing
X_resampled_df = pd.DataFrame(X_resampled, columns=X.columns)
X_resampled_df['default'] = y_resampled

print("\nFirst few rows of resampled data:")
print(X_resampled_df.head())
```
